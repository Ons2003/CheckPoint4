```python
#QUESTION 1 
def Maximum(a,b,c) : 
  if a<b :
    max = b
  else :
    max = a 
  if max < c :
    max = c 
  return max 
a, b, c = input("Enter three values: ").split()
print(Maximum(a,b,c))
```

```python
#QUESTION 3 
l = [3,7,6,5,8]
  #Summation Function
def Sum(list0) :
  s = 0 
  for i in list0 : 
    s = s+i
  return s
  #Product Function 
def Prod(list0) :
  p = 1 
  for i in list0 :
    p = p*i
  return p 
  #Extraction
new_list1 = []
new_list2 = []
for i in range(0,len(l),2) :
  new_list1.append(l[i])
for i in range(1,len(l),2) :
  new_list2.append(l[i])

print(Sum(new_list1))
print(Prod(new_list2))
```

```python
#QUESTION 5 
import math 
n = input('Give a sequence of numbers seperated by a coma')
def Bonus(ch) :
  Q = ''
  C = 50 
  H = 30
  l= ch.split(',')
  for i in l :
    D = int(i)
    Q = Q + str(int(math.sqrt( ((2 * C * D)/H) ))) + ','
  return Q
print(Bonus(n))
```
